**The Ising model**

_Refer to Exercise 10.9 in the Newman text._ The Ising model is a theoretical model of a magnet. The magnetic potential energy due to the interaction of two dipoles is proportional to their dot product, but in the Ising model this simplifies to just the product $s_i s_j$ for spins on sites $i$ and $j$ of the lattice, since the spins are one-dimensional scalars, not vectors. Then the actual energy of interaction is $-Js_i s_j$, where $J$ is a positive interaction constant. The minus sign ensures that the in- teractions are ferromagnetic, meaning the energy is lower when dipoles are lined up. A ferromagnetic interaction implies that the material will magnetize if given the chance.

Normally it is assumed that spins interact only with those that are immediately adjacent to them on the lattice, which gives a total energy for the entire system equal to
$$ E = -J \sum_{\langle ij \rangle} s_i s_j,$$
where the notation $\langle ij \rangle$ indicates a sum over pairs $i,j$ that are adjacent on the lattice. On the square lattice we use in this exercise each spin has four adjacent neighbors with which it interacts, except for the spins at the edges of the lattice, which have either two or three neighbors.

Perform a Markov chain Monte Carlo simulation of the Ising model on the square lattice for a system of 20 x 20 spins. You will need to set up variables to hold the value ±1 of the spin on each lattice site, probably using a two-dimensional integer array, and then take the following steps.

1. Write a function to calculate the total energy of the system, as given by the equation above. That is, for a given array of values of the spins, go through every pair of adjacent spins and add up the contributions $s_i s_j$ from all of them, then multiply by $-J$. Hint 1: Each unique pair of adjacent spins crops up only once in the sum. Thus there is a term $-Js_1s_2$ if spins 1 and 2 are adjacent to one another, but you do not also need a term $-Js_2s_1$. Hint 2: To make your final program to run in a reasonable amount of time, you will find it helpful if you can work out a way to calculate the energy using Python's ability to do arithmetic with entire arrays at once. If you do the calculation step by step, your program will be significantly slower. *(25 pts.)*

2. Now use your function as the basis for a Metropolis-style simulation of the Ising model with $J = 1$ and temperature $T = 1$ in units where the Boltzmann constant $k_B$ is also 1. Initially set the spin variables randomly to ±1, so that on average about a half of them are up and a half down, giving a total magneti- zation of roughly zero. Then choose a spin at random, flip it, and calculate the new energy after it is flipped, and hence also the change in energy as a result of the flip. Then decide whether to accept the flip using the Metropolis acceptance formula, Eq. (10.60). If the move is rejected you will have to flip the spin back to where it was. Otherwise you keep the flipped spip. Now repeat this process for many moves. *(25 pts.)*

3. Make a plot of the total magnetization $M = \sum_i s_i$ of the system as a function of time for a million Monte Carlo steps. You should see that the system develops a "spontaneous magnetization," a nonzero value of the overall magnetization. Hint: While you are working on your program, do shorter runs, of maybe ten thousand steps at a time. Once you have it working properly, do a longer run of a million steps to get the final results. *(25 pts.)*

4. Run your program several times and observe the sign of the magnetization that develops, positive or negative. Describe what you find and give a brief explana- tion of what is happening. *(25 pts.)*

```python
#--------------------------------A. Imports
import numpy as np
from numpy import array, random
from math import exp
import matplotlib.pyplot as plt

#--------------------------------I. Energy Function (n-dimensional)
def spinenergy(lattice, J=1):               # calculates spin energy by 'dotting' adjacent vectors [NOTE: see Appendix B for full explanation]
    energy = 0
    for d in range(np.ndim(lattice)):
        lattice = np.swapaxes(lattice,0,d)
        energy += np.sum((lattice[1:,...] * np.roll(lattice,1,0)[1:,...]))      
    return -J*energy

#------------------------------II.1 Lattice Generator
def latticegen(size = 20,depolarize = False):
    states = [-1,1]                                     # given a list of possible states, we generate our lattice array with the spins being equally likely to be in any of the states
    if type(size) is int:                               # by DEFAULT, we generate a 2D-array of shape (size, size), where *size* is by DEFAULT set to 20. 
        lattice = random.choice(states,[size,size])     
    else: 
        lattice = random.choice(states,size)            # if user inputs an array, however, we generate an array of shape *size*, each with a randomly set state.          
    
    if depolarize == True:                                                      # OPTIONAL: if arg *depolarize* is True, 
        while abs(np.sum(lattice)) > 0:                                         # latticegen() randomly flips spins until total Magnetization is zero. [NOTE: See Appendix A]
            randspin = tuple([random.randint(0, i) for i in np.shape(lattice)]) 
            lattice[randspin] = -1 * np.sum(lattice)/abs(np.sum(lattice))        
    return lattice

#------------------------------II.2 Metropolis Model Step
kB=1              # Boltzmann's constant k_B
J = 1       
T = 1
beta = 1/(kB*T)   # for convenience, we define β = (k_B*T)^{-1}

def metrostep(lattice, J = 1):         
    E = spinenergy(lattice,J)
    yield lattice                  # first step just yields the input lattice
    while True:                                                             # NOTE: our generator has no option to terminate.
        randspin = tuple([random.randint(0, i) for i in np.shape(lattice)]) # we create a random n-tuple... 
        lattice[randspin] *= -1                                             # ...which will be the 'coordinate' of the spin we flip
        Enew = spinenergy(lattice,J)  # energy Enew of lattice after flip
        Pa = exp(-beta*(Enew-E))      # Probability of acceptance
        if random.rand() < Pa:                                              # we create a condition with probability *P_a* of succeeding, i.e. generating a number [0, 1) and seeing if it's greater than P_a.
            E = Enew                                                        # if we accept, then the lattice stays changed, and we replace E with our new lattice's energy Enew
        else:
            lattice[randspin] *= -1                                     # if we reject, we undo the change we made to the lattice
        yield lattice                                                   # we yield the lattice we end up with (which will be our 'current' configuration)
```

For $10^6$ Monte Carlo steps, we have the following graph for Total Magnetization of a randomly-generated lattice:

```python
steps = 1000000
metromodel = metrostep(latticegen())
totalmag = [np.sum(next(metromodel)) for _ in range(steps)]
plt.plot(totalmag)
```

Repeating this a few more times, we have:

Here, we can see that after enough steps, our randomly-generated lattices always end up with a spontaneous magnetization, which is either negative or positive. These results can be explained in two parts:

### I. Probability of Flipping
From the formula for energy, we see that $\sum s_i s_j$ **increases** whenever we have more similar terms, with each opposite-pair $1*(-1)$ having a **negative** contribution to the sum. Since energy is given by $E=-J\cdot \sum s_i s_j$, energy should then **decrease** with more similar terms.

To visualize this easier, we consider the $3\times 3$-matrix around some point $lattice[i,j],$ since our change in energy can only come from this neighborhood $[!]$. In the 4 points adjacent to $lattice[i,j]$, we either have equal amounts of both states $1,-1$ or more of one state. In the first case, flipping the center leads to the same energy for this matrix, $e=-Js_{ij}(s_{i-1,j} + s_{i+1,j} + s_{i,j-1}+s_{i,j+1})=\mp Js_{ij}(2-2)=0.$ In the second case, flipping the center to the more common state results in a **lower** energy, which gives a **negative** $\Delta E.$ This again follows from the above formula, where flipping now results in more products being equal to 1. Lastly, flipping the center to the less common state results in a **higher** energy and **positive** $\Delta E.$

In our case where the states are $[-1,1]$, if there were $n$ spins in the more dominant state, flipping to that state yields in a change of $(n-2^d)-(2^d-n)=2n$ in the sum of products, i.e. a **change-in-energy** $\Delta E =-J\cdot 2n,$ where the dimension $d$ doesn't change this fact. Similarly, flipping to the less dominant state results in $\Delta E = J\cdot 2n.$ In the first two scenarios, we have $\Delta E\leq 0,$ which results in a probability $P_a=\exp(-k), \ k \leq 0,$ i.e. $P_a = 1$ when $k=0$ and $P_a > 1$ otherwise. So, such flips are *always* accepted by our model. In the last scenario, we have $P_a < 1,$ meaning they're accepted less often (since they're not guaranteed). 

$[!]$ See Appendix $C$ for an alternate function $\text{metrostep2()}$ that takes advantage of this.

### II. Total Magnetization Over Time
At first, our lattice has roughly neutral magnetization. And while our flips are already biased towards the "more common" state, **note** that this only looks at the adjacent 4 states, meaning it's equally likely that one or the other state is "more common" in the neighborhood of our randomly-selected spin. So, on average, we're equally likely to increase the number of spins in either of the states.

However, our lattices aren't completely neutral - they may start with more of one state than the other. And even if they were neutral initially, each step **always** results in **one more** spin having one of the states and **one less** spin having the other. So, we always have more of one state than the other in our system, either in our **current timestep** or, if not, in the timestep **right after**.

Given this, there is **always** one state that, on average, will be "more common" in the neighborhood around some randomly-selected spin. Once the system flips more of the spins to this state, then the likelihood of flipping more spins to this state increases, and the number of spins in this state increases rapidly. And if the system reverts back to neutral, then the next timestep will give it non-neutral magnetization, and there is again a chance that the system flips more spins to the dominant state. So, given enough timesteps, we will **always** end up with one of the states being much more common than the other. In our test, they hover at $\sum_{i,j} s_{ij}=\pm 400,$ i.e. with all states being either $+1$ or $-1,$ which is consistent with our logic above. 

Even after reaching this state, there is still a non-zero probability of flipping a spin to the less common state (from the equation) — and if we increased the resolution of our graphs, we **would** see that this is the case. However, this is now maximally improbable: any randomly-selected spin will have all its adjacent spins be in the same state as it, meaning the change-in-energy that comes from flipping it would be at its greatest (positive) value. Given dimension $d,$ we have:

$$\begin{align*}
\Delta E &= J \cdot 2n, \ n = 2^d \\
&= J \cdot 2^{d+1}\\
P_a &= \exp\left(-\frac{1}{k_B T} \cdot \Delta E\right) \\
 &= \exp\left(-\frac{J}{k_B T} \cdot 2^{d+1}\right) \ \ ★
\end{align*}$$

From our equations here, we notice that $|J|$ plays the role of minimizing $P_a.$ However, note that this doesn't change the sign of the exponent — thus, $P_a \geq 1$ will still be true when flipping to the more dominant state, and this change is only felt when attempting to flip to the less dominant state. So, increasing $|J|$ increases the probability of **rejecting** flips to the less dominant state, which should make the lattice converge faster to its final state:

```python
steps = 1000000
lattice = latticegen()

lattice1 = lattice.copy()
metromodel1 = metrostep(lattice1)

J = 20
lattice2 = lattice.copy()
metromodel2 = metrostep(lattice2, J)

totalmag1 = [np.sum(next(metromodel1)) for _ in range(steps)]
totalmag2 = [np.sum(next(metromodel2)) for _ in range(steps)]
plt.plot(totalmag1)
plt.plot(totalmag2)
```

where here, the Blue graph (by default) represents the first attempt (which was done with $J = 1$) and the Orange one represents the attempt with $J = 20.$ Note also that the **sign** of $J$ changes the expected final state: since this amounts to 'swapping' our logic on the probabilities, having $J<0$ means flips to the **less dominant** state are always accepted, and the flips to the more dominant state will be the ones accepted with probability $P_a < 1.$ 

So, 'reversing' our logic in Section $II$ above, if one of the states were more likely to be the dominant one in the neighborhood of a randomly-selected spin, then we're **more likely** to flip to the less dominant state, i.e. **converting** spins from the more dominant state to the less dominant state. This means we should expect our Total Magnetization to converge towards $0,$ i.e. neutral:

```python
J = -30     # a LARGE J (quickly converging), with negative sign (converging towards checkerboard)
steps = 100000
lattice3 = latticegen()
metromodel3 = metrostep(lattice3, J)
totalmag3 = [np.sum(next(metromodel3)) for _ in range(steps)]
plt.plot(totalmag3)
```

 **Note** however that this will only be stable if our state is such that any randomly-selected spin is **very likely** to be surrounded by more spins in the opposite state - or, in other words, when our energy is smallest (since at that point, flips can only increase energy). This means that the "**Checkerboard**"-configuration is the most stable one, since there **every** spin is surrounded by spins in the opposite state:

```python
iterations = 100
latticehistory = [next(metromodel3) for _ in range(iterations)]         # We see how the lattice evolves for 100 iterations
latticeavg = np.copy(latticehistory[0])                                 
for k in range(1, iterations):
    latticeavg += latticehistory[k]                                     # Then, we SUM all the lattices we've stored...
latticeavg = latticeavg/100                                             # and divide by number of iterations to get the AVERAGE state it's in.
plt.imshow(latticeavg,cmap='gray')
```

Finally, by the same logic, we see that $T$ has a similar effect, noting that $T\geq0$ here (in Kelvin). Since it's in the denominator, increasing $T$ has the same effect as decreasing $|J|,$ i.e. it makes the system **take more time** to reach its stable state. In particular, it *increases* the probability of the non-guaranteed action. 

At first, it seems like this should imply that the stable states for $J>0$ and $J<0$ get "swapped": with the $J>0$ case, for example, the system now has a greater probability of flipping to the **less dominant** state. *However*, note that this again just affects our results **when** $P_a < 1$ - that is, in both cases, flipping to the more favorable state is **still always done**, only this time flipping to the less favorable state is now more probable as well, approaching $1$ (i.e. guaranteed) the greater our $T$ becomes. 

In both cases then, increasing $T$ has the effect of keeping the system at its initial Total Magnetization. This is easier to imagine for very high $T:$ in this case, we'd be flipping spins almost always, regardless of their neighboring spins. **Note** that this doesn't mean both cases approach the Checkerboard-state: in fact, since flipping becomes more indiscriminate for higher $T$'s, our system's configuration (which was generated randomly) will remain random, since we're just editing randomly.

##### B. The $\text{spinenergy()}$ Function

According to Newman, the energy of the system is a sum of the **products** of adjacent spins, $\sum s_i s_j$ multiplied by a $-J$ factor. He *notes* that we only count adjacent spins once (it'd be double-counting otherwise) and **more importantly**, that the term "adjacent" excludes spins that are a diagonal step away from each other. In the 2D-case, this means that given some $s_{ij}$, the terms adjacent to it are $s_{i-1,j}$ and $s_{i+1,j}$ along the first axis and $s_{i,j-1}$ and $s_{i,j+1}$ along the second axis. To count the sum of their products **once**, we can cover those along the first axis by dotting adjacent rows and those along the second axis by dotting adjacent columns. Since direction doesn't matter, we do this left-to-right (along each given axis), i.e. dot $s_{i,\alpha}$ with $s_{i+1,\alpha}$ which stops when we dot the second-to-last row/column with the last one.

This **more easily generalizes** to the $n$-dimensional case thanks to our definition of "adjacent": while there are many subsets of arrays that we can choose to dot with each other (i.e. multiply element-wise), there are only $n$ **senses in which terms can be adjacent** - i.e., one for each axis we have. So, we can already 'capture' the total contribution due to adjacent spins just by dotting the adjacent $n-1$-dimensional arrays along each axis. Furthermore, along any given axis, the order in which we dot adjacent vectors and sum their contributions doesn't matter. In the 3D-case, for example, if we wish to add the contributions along the "$x$"-axis, dotting the vectors along $y$ first then summing across $z$ leads to the same sum as dotting then summing the reverse order, since we're essentially summing the element-wise products of the same two "faces" (the $yz$-plane in the $i$-th and $i+1$-th slot) with each other. 

Below is an implementation of the above logic:

```python
def spinenergy_old(lattice, J=1):                                   
    energy = 0
    for d in range(np.ndim(lattice)):                               # Looks at each axis...
        lattice=np.swapaxes(lattice,0,d)                            # puts this axis "first"...
        for k in range(lattice.shape[0] - 1):                       # then iterates over the (k-1) adjacent (n-1)-dimensional arrays along it...
            energy += np.sum(lattice[k,...] * lattice[k+1,...])     # and dots adjacent arrays, adding its contribution to the total energy.
    return -J*energy
```

In the above, we put our current axis "first" in our view of the array via $\text{swapaxes()}$. This is because in array indexing, we can use the term "$...$" to indicate that for all succeeding axes, we have "$:$" in their slots, i.e. we include all elements along these axes. However, to have it input "$:$" for **all** the other slots, we'd have to put the current axis in front. Explicitly, our axes are now arranged as: $[k,0,1,...,k-1,k+1,...,n].$ Again, the ordering of our axes won't matter as long as we're dotting the same $n-1$-dimensional arrays with each other.

The above implementation, however, can be improved by using the **roll()** function, which shifts over our $n-1$-dimensional arrays forward, looping them back to the beginning whenever they're already at the last slot. This is more inexpensive because instead of dotting arrays one pair at a time, we can dot all $k-1$ pairs simultaneously, which is faster due to **numpy**'s implementation. Explicitly, in our final implementation, we shift our lattice array once forward along the $d$-th axis, then dot the second-until-last arrays (i.e. $[1:]$) of the shifted and unshifted lattice arrays, which amounts to dotting the 1st to the 2nd, 2nd to 3rd, etc.


For larger-sized arrays, this effect is significant even in the 2D-case — in my device the energy of a 2D lattice with $n=20,000$ can be computed in $5.6s$ instead of $11.6s$:

##### C. Alternate $\text{metrostep()}$-function

As mentioned in Section $I,$ all that contributes to $\Delta E$ is the sum of products from pairs **adjacent to** the randomly-selected spins which we will consider flipping. If our model **by design** knows which spins are to be flipped, then we can take advantage of this in our $\text{metrostep()}$ function:

```python
def metrostep2(lattice, J = 1):
    yield lattice                  
    while True:                                                             
        randspin = [random.randint(0, i) for i in np.shape(lattice)]  
        dE = 0
        for d in range(np.ndim(lattice)):                                   # we compute ΔE in EACH dimension
            left = randspin.copy()                                          # we create coordinates *left* and *right* which are EQUAL to *randspin* at first
            right = randspin.copy()
            if randspin[d] != 0:                                            # if *randspin*'s d-coordinate isn't the leftmost, then there exists a spin TO THE LEFT.
                left[d] -= 1
                dE += -2*lattice[tuple(randspin)]*lattice[tuple(left)]      # we add ΔE_left = (-1*s_rand)*s_left - s_rand*s_left = -2s_rand*s_left

            if randspin[d] != np.shape(lattice)[d] - 1:                     # if *randspin*'s d-coordinate isn't the rightmost, then there exists a spin TO THE RIGHT.
                right[d] += 1   
                dE += -2*lattice[tuple(randspin)]*lattice[tuple(right)]     # same as in the *left* case
                
        dE *= -J                                                        # We have ΔE = -J*(ΔΣ)
        # NOTE: "TEST" code was inserted here to confirm that this method returns the same energies as spinenergy(lattice)                
        Pa = exp(-beta*(dE))      
        if random.rand() < Pa:                                              
            lattice[tuple(randspin)] *= -1                                     # if we accept
        yield lattice                                                   # we yield the lattice we end up with (which will be our 'current' configuration)
```

In the above, we compute $\Delta E$ *directly* by just summing the products of the chosen spin with those adjacent to it, and keep the rest of the code from the original $\text{metrostep()}$ function. As can be seen below, it's significantly faster:

Where our $\text{metrostep2()}$ function takes roughly **one-third** of the time to run a million Monte Carlo steps. The reason Newman doesn't do so might be that in practice, the model wouldn't actually know them beforehand, which requires us to look at the initial and final energies of our **entire** system. In this case, it would then be a question of whether calculating the $\Delta E$ of the **entire system** is cheaper than searching for the flipped spins then calculating the $\Delta E_i$'s that they contribute. To search for flipped spins, we could subtract the lattice **after the flip** from the initial lattice and then search for non-zero elements. 

Since our lattice is random, searching for just one flipped spin amounts to searching through each spin, with a probability $1/K$ of finding it on the $k$-th spin, where $K$ is the total number of spins. On average, this would mean conducting $(1+2+3+...+K)/K=(K+1)/2$ searches, meaning that for an $n$-dimensional lattice searching for a flipped spin runs in $O(n^d).$ **Note** that searching for $x$ flips would still run in the same time, since we'd just note where the flipped spins are located each time. This would be *much slower* than just dotting entire arrays together, thanks to numpy's implementation.
